```python
from detectron2.engine import DefaultTrainer
from detectron2.evaluation import COCOEvaluator

class CocoTrainer(DefaultTrainer):

  @classmethod
  def build_evaluator(cls, cfg, dataset_name, output_folder=None):

    if output_folder is None:
        os.makedirs("coco_eval", exist_ok=True)
        output_folder = "coco_eval"

    return COCOEvaluator(dataset_name, cfg, False, output_folder)
```

```python
import optuna
import copy
from detectron2.config import get_cfg
from detectron2.data import build_detection_test_loader
from detectron2.evaluation import inference_on_dataset

cfg = get_cfg()
cfg.merge_from_file(model_zoo.get_config_file("COCO-Detection/faster_rcnn_R_50_FPN_3x.yaml"))
cfg.DATASETS.TRAIN = ("my_dataset_train",)
cfg.DATASETS.TEST = ("my_dataset_val",)
cfg.MODEL.WEIGHTS = model_zoo.get_checkpoint_url("COCO-Detection/faster_rcnn_R_50_FPN_3x.yaml")

def objective_fasterrcnn(trial):
    cfg_opt = copy.deepcopy(cfg)

    cfg_opt.SOLVER.BASE_LR = trial.suggest_float("base_lr", 0.0001, 0.001, log=True)
    cfg_opt.SOLVER.WARMUP_ITERS = trial.suggest_int("warmup_iters", 500, 1000)
    cfg_opt.MODEL.ROI_HEADS.BATCH_SIZE_PER_IMAGE = trial.suggest_categorical("roi_batch_size", [64, 128])

    cfg_opt.SOLVER.MAX_ITER = 1500
    cfg_opt.SOLVER.IMS_PER_BATCH = 16
    cfg_opt.SOLVER.STEPS = (1000,1200)
    cfg_opt.SOLVER.GAMMA = 0.1
    cfg_opt.INPUT.MIN_SIZE_TRAIN = (640,)
    cfg_opt.INPUT.MAX_SIZE_TRAIN = 640
    cfg_opt.INPUT.MIN_SIZE_TEST = 640
    cfg_opt.INPUT.MAX_SIZE_TEST = 640
    cfg_opt.DATALOADER.NUM_WORKERS = 4
    cfg_opt.MODEL.ROI_HEADS.NUM_CLASSES = 9
    cfg_opt.TEST.EVAL_PERIOD = 1000
    cfg_opt.SOLVER.CHECKPOINT_PERIOD = 1000
    cfg_opt.OUTPUT_DIR = f"/content/output/trial_{trial.number}"
    cfg_opt.MODEL.DEVICE = "cuda"

    os.makedirs(cfg_opt.OUTPUT_DIR, exist_ok=True)

    trainer = CocoTrainer(cfg_opt)
    trainer.resume_or_load(resume=False)
    trainer.train()

    evaluator = COCOEvaluator("my_dataset_val", cfg_opt, False, output_dir=cfg_opt.OUTPUT_DIR)
    val_loader = build_detection_test_loader(cfg_opt, "my_dataset_val")
    results = inference_on_dataset(trainer.model, val_loader, evaluator)

    return results["bbox"]["AP50"]
```

```python
study = optuna.create_study(
    study_name="fasterrcnn_tuning",
    storage=db_path,
    direction="maximize",
    load_if_exists=True
)
study.optimize(objective_fasterrcnn, n_trials=1)

print("Best trial:")
print(study.best_trial.params)
```
